```python
# limit, interval
nrows = 2
ncols = 2
nfigs = nrows*ncols

fig, ax = plt.subplots(
    nrows, ncols, 
    sharex=True, sharey=False, 
    figsize=(cm2in(21.5),cm2in(13*2)), 
    facecolor='w', edgecolor='k',
    squeeze=False,
    dpi = 600,
    )

# tick settings 
ymin = 120
ymax = 144
yint = 6
cmax = 0.42

xlabel = ["Elapsed time [h]"]*3

color_list = ['#F7F7F7','#9737A7'] 
colormap1 = mcolors.LinearSegmentedColormap.from_list(name = "colormap",colors=color_list, N=300)

# 플랏 그리기
for ridx in range(nrows):
    for cidx in range(ncols): 
        
        # figure index 
        idx = ncols*ridx + cidx

        N = DF[idx].shape[1]
        norm = mcolors.Normalize(vmin = 0, vmax = cmax)

        im = ax[ridx,cidx].imshow(
        DF[idx],
        cmap='BuPu', # Purples2, Sunset, Sunrise, magma
        norm=norm,
        extent=[0, N, ymax, ymin],
        aspect='auto',
        interpolation='none'
        )
        
        xmar = 0.5
        ymar = 0.5

        ax[ridx, cidx].grid(False)

        ax[ridx,cidx].set_xlim(0-xmar, DF[idx].shape[1]+xmar)
        ax[ridx,cidx].set_ylim(ymin-ymar, ymax+ymar)
        ax[ridx,cidx].invert_yaxis()

        ax[1,cidx].set_xlabel('x coordinate [cm]', fontsize =fs)
        ax[ridx,0].set_ylabel('Elapsed time [h]', fontsize = fs)
        
        # annotation
        subplot_no = list(map(chr,range(97,123)))
        case_name = [r'US ($\mathit{\alpha}= 0.5\times 10^{-6}~\mathrm{m^2/s}$)',
                     r'US ($\mathit{\alpha}= 1.0\times 10^{-6}~\mathrm{m^2/s}$)',
                     r'US ($\mathit{\alpha}= 2.0\times 10^{-6}~\mathrm{m^2/s}$)',
                     'D']
        subplot_idx = '('+subplot_no[idx]+') ' + case_name[idx] #(a) + Internal
        ax[ridx,cidx].annotate(subplot_idx, xy=(.01, 1.045), xycoords='axes fraction',
            horizontalalignment='left', verticalalignment='top', fontsize=Annotation_FS) 
        
        # 축 눈금 및 라벨 크기 설정
        ax[ridx,cidx].tick_params(direction='in', labelsize=Axis_FS, which='major', length = 2.5, width=0.5, pad = 3.5)
        ax[ridx,cidx].tick_params(direction='in', labelsize=Axis_FS, which='minor', length = 1.25, width=0.25, pad = 3.5)

        ax[ridx,cidx].set_xticks(np.arange(0, DF[idx].shape[1]+5, 5)) 
        ax[ridx,cidx].set_yticks(np.arange(ymin, ymax+yint, yint)) 

        ax[ridx,cidx].xaxis.set_minor_locator(ticker.AutoMinorLocator(5))
        ax[ridx,cidx].yaxis.set_minor_locator(ticker.AutoMinorLocator(3))
                # spine line width  
        for k in ['top','bottom','left','right']:
                ax[ridx,cidx].spines[k].set_visible(True)
                ax[ridx,cidx].spines[k].set_linewidth(0.5)
                ax[ridx,cidx].spines[k].set_color('k') 

plt.subplots_adjust(wspace=0.18,hspace=0.12, 
                    left=0.07,
                    right=0.93,
                    bottom=0.025,
                    top=0.975)
fig.subplots_adjust(bottom=0.13)
# fig.tight_layout(h_pad=5, w_pad=5)

# colorbar
cbar_width =  0.015 #vertical  
cbar_height = 0.010 #horizontal  
cbar_dist_v = 0.06; # vertical colorbar distance from bbox edge
cbar_dist_h = 0.03; # horizontal colorbar distance from bbox edge

# horizontal colorbars 
# colorbar1 temperature 
bbox1 = ax[1,0].get_position() # get the normalized axis original position 
bbox3 = ax[1,1].get_position() # get the normalized axis original position 
cb_ax1 = fig.add_axes([bbox1.x0, bbox1.y0-cbar_dist_v, bbox3.x1-bbox1.x0, cbar_height]) #[x_origin, y_origin, width, height]
cbar1  = fig.colorbar(im, cax=cb_ax1, ax=ax[0,0], orientation='horizontal') 
minor_locator = ticker.MultipleLocator(0.05)
cbar1.ax.xaxis.set_minor_locator(minor_locator)
cbar1.ax.tick_params(which = "major" ,direction='in',labelsize=Axis_FS, length = 3, width = 0.5)
cbar1.ax.minorticks_off()
# cbar1.locator = ticker.MaxNLocator(nbins=10)
cbar1.locator = ticker.MultipleLocator(0.06)
cbar1.ax.set_xlabel('Exergy consumption rate [W/m$^2$]', fontsize=FS_label)
cbar1.update_ticks()
cbar1.outline.set_linewidth(0.5)

# figname = 'Fig8_StoredExergyComparisonHeatmap'
# plt.savefig(figname+'.svg', dpi=600)
# plt.savefig(figname+'.png', dpi=600)
# plt.savefig(figname+'.pdf', dpi=600)
```
